```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["survived"].value_counts(normalize=True) * 100
```

```python
sns.barplot(x="sex", y="survived", data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x="class", y="survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 12, 18, 35, 60, 100],
    labels=["Child", "Teen", "Young Adult", "Adult", "Senior"]
)

df["age_group"].value_counts()
```

```python
sns.barplot(x="age_group", y="survived", data=df)
plt.title("Survival Rate by Age Group")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x="survived", y="age", data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

## Insight Summary

- Females had significantly higher survival rates compared to males.
- First-class passengers were more likely to survive than third-class passengers.
- Children showed relatively higher survival rates.
- Age distribution indicates younger passengers had better survival probability.
- Socioeconomic status (class) played a major role in survival chances.
